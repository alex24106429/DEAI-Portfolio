 Opdracht 1:

 Schrijf een (recursieve) methode ’subList(self, start, end)’, die een nieuwe lijst retourneert met alle elementen die

 in de oorspronkelijke lijst tussen positie ’start’ (inclusief) en ’end’ (exclusief) liggen. Een aanroep bijvoorbeeld van

 ’subList(1, 3)’ op een lijst met elementen ’5’, ’4’, ’7’ en ’4’ levert een nieuwe lijst op met elementen ’4’ en ’7’. Schrijf

 tests die de werking van de methode aannemelijk maken

```python
import csv
import sys


class LinkedList:
    pass


class LinkedListEmpty(LinkedList):
    def __init__(self):
        pass

    def toString(self):
        return ""

    def addFirst(self, value):
        return LinkedListPopulated(value, self)

    def subList(self, start, end):
        return LinkedListEmpty()


class LinkedListPopulated(LinkedList):
    def __init__(self, head, tail):
        self.head = head
        self.tail = tail

    def toString(self):
        return str(self.head) + " " + self.tail.toString()

    def addFirst(self, value):
        return LinkedListPopulated(value, self)

    def subList(self, start, end):
        if end <= 0:
            return LinkedListEmpty()
        if start > 0:
            return self.tail.subList(start - 1, end - 1)
        return LinkedListPopulated(self.head, self.tail.subList(0, end - 1))


test_list = LinkedListEmpty().addFirst(4).addFirst(7).addFirst(4).addFirst(5)
print(test_list.subList(1, 3).toString())
```

 Opdracht 3: Schrijf een (recursieve) methode ’merge(self, list)’, die een nieuwe lijst retourneert met alle elementen uit de oor-

 spronkelijke en de gegeven lijst samengevoegd en gesorteerd op volgorde van klein naar groot. Er mag vanuit gegaan

 worden dat de oorspronkelijke en de gegeven lijst beide gesorteerd zijn van klein naar groot. Een aanroep bijvoor-

 beeld van ’merge(list)’ op een lijst met elementen ’4’, ’4’, ’5’ en ’7’ en waarbij ’list’ een lijst is met elementen ’2’, ’6’

 en ’7’ levert een nieuwe lijst op met elementen ’2’, ’4’, ’4’, ’5’, 6’, ’7’ en ’7’. Bepaal zelf in hoeverre je van recursie

 en/of hulpmethoden gebruik wilt maken. Schrijf tests die de werking van de methode aannemelijk maken.

```python


def merge_empty(self, andere_lijst):
    return andere_lijst


LinkedListEmpty.merge = merge_empty


def merge_populated(self, andere_lijst):
    if isinstance(andere_lijst, LinkedListEmpty):
        return self
    if self.head <= andere_lijst.head:
        return LinkedListPopulated(self.head, self.tail.merge(andere_lijst))
    else:
        return LinkedListPopulated(andere_lijst.head, self.merge(andere_lijst.tail))


LinkedListPopulated.merge = merge_populated


def run_tests():
    lijst1 = LinkedListEmpty().addFirst(7).addFirst(5).addFirst(4).addFirst(4)
    lijst2 = LinkedListEmpty().addFirst(7).addFirst(6).addFirst(2)

    resultaat1 = lijst1.merge(lijst2)

    assert resultaat1.toString(
    ) == "2 4 4 5 6 7 7 ", f"Fout in Test: {resultaat1.toString()}"

    print("De test is succesvol doorlopen!")


run_tests()
```

 Opdracht 6: Schrijf een (recursieve) methode ’sortMerge()’, die een nieuwe lijst retourneert met alle elementen uit de oorspronke-

 lijke lijst gesorteerd van klein naar groot. Een aanroep van ’sortMerge()’ op een lijst met elementen ’5’, ’4’, ’7’ en ’4’

 bijvoorbeeld levert een nieuwe lijst op met elementen ’4’, ’4’, ’5’ en ’7’. Schrijf tests die de werking van de methode

 aannemelijk maken.

```python


def length_empty(self):
    return 0


LinkedListEmpty.length = length_empty


def length_populated(self):
    return 1 + self.tail.length()


LinkedListPopulated.length = length_populated


def sortMerge_empty(self):
    return self


LinkedListEmpty.sortMerge = sortMerge_empty


def sortMerge_populated(self):
    length = self.length()
    if length == 1:
        return self

    mid = length // 2

    left_half = self.subList(0, mid)
    right_half = self.subList(mid, length)

    sorted_left = left_half.sortMerge()
    sorted_right = right_half.sortMerge()

    return sorted_left.merge(sorted_right)


LinkedListPopulated.sortMerge = sortMerge_populated

lijsttestmerge = LinkedListEmpty().addFirst(
    4).addFirst(7).addFirst(4).addFirst(5)
mergedlist = lijsttestmerge.sortMerge()
print(mergedlist.toString())
```
